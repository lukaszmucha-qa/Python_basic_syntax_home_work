```python
# Zadanie 1 - Kalkulator BMI
waga = float(input("Podaj swoją wagę w kilogramach: "))
wzrost = float(input("Podaj swój wzrost w metrach: "))
bmi = waga / (wzrost ** 2)
print(f"BMI: {bmi:.2f}")
if bmi < 18.5:
    print("Niedowaga")
elif 18.5 <= bmi < 24.9:
    print("Waga prawidłowa")
elif 25 <= bmi < 29.9:
    print("Nadwaga")
else:
    print("Otyłość")


```

```python
#Zadanie 2 - Konfersja temperatur
temp_c = float(input("Podaj temperaturę w stopniach Celsjusza: "))
temp_f = (temp_c * 9/5) + 32
print(f"Temperatura w stopniach Fahrenheita: {temp_f:.2f}")
temp_k = temp_c + 273.15
print(f"Temperatura w stopniach Kelvina: {temp_k:.2f}")
print(f"{temp_c:.2f}°C = {temp_f:.2f}°F = {temp_k:.2f}K")
```
