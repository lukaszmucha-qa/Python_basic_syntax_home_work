```python
# Zadanie 1 - Kalkulator BMI
weigh = float(input("Podaj swoją wagę w kilogramach: "))
height = float(input("Podaj swój wzrost w metrach: "))
bmi = weigh / (height ** 2)
print(f"BMI: {bmi:.2f}")
if bmi < 18.5:
    print("Niedowaga")
elif 18.5 <= bmi < 24.9:
    print("Waga prawidłowa")
elif 25 <= bmi < 29.9:
    print("Nadwaga")
else:
    print("Otyłość")


```

```python
# Zadanie 13 (Listy) - Analiza Iris Dataset
# Notatki do zadania 13
# Iris dataset - uproszczona wersja (5 pierwszych rekordów)
# Format: sepal_length, sepal_width, petal_length, petal_width, species
iris_data_raw = """5.1,3.5,1.4,0.2,setosa
4.9,3.0,1.4,0.2,setosa
4.7,3.2,1.3,0.2,setosa
4.6,3.1,1.5,0.2,setosa
5.0,3.6,1.4,0.2,setosa
7.0,3.2,4.7,1.4,versicolor
6.4,3.2,4.5,1.5,versicolor
6.9,3.1,4.9,1.5,versicolor
5.5,2.3,4.0,1.3,versicolor
6.5,2.8,4.6,1.5,versicolor
6.3,3.3,6.0,2.5,virginica
5.8,2.7,5.1,1.9,virginica
7.1,3.0,5.9,2.1,virginica
6.3,2.9,5.6,1.8,virginica
6.5,3.0,5.8,2.2,virginica"""
iris_data_lines = iris_data_raw.split('\n')
data_as_list = []
list_of_species = []
list_of_measurements_per_species = []
for line in iris_data_lines:
    data_as_list = line.split(',')
    data_as_list[0] = float(data_as_list[0])
    data_as_list[1] = float(data_as_list[1])
    data_as_list[2] = float(data_as_list[2])
    data_as_list[3] = float(data_as_list[3])
    data_as_list[4] = data_as_list[4]
    list_of_measurements_per_species.append(data_as_list)
    list_of_species.append(data_as_list[4])
list_of_species = list(set(list_of_species))

avg_measurements_per_species =[]
max_petal_width_per_species = []
petal_length_to_width_ratio_per_species = []
for species in list_of_species:
    petal_width_measurements = []
    petal_length_measurements = []
    for measurement in list_of_measurements_per_species:
        if measurement[4] == species:
            petal_length_measurements.append(measurement[2])
            petal_width_measurements.append(measurement[3])
    avg_petal_length = round(sum(petal_length_measurements) / len(petal_length_measurements) , 2)
    avg_measurements_per_species.append((species, avg_petal_length))
    max_petal_width = max(petal_width_measurements)
    max_petal_width_per_species.append((species, max_petal_width))
    avg_petal_length = sum(petal_length_measurements) / len(petal_length_measurements)
    avg_petal_width = sum(petal_width_measurements) / len(petal_width_measurements)
    ratio = round(avg_petal_length / avg_petal_width)
    petal_length_to_width_ratio_per_species.append((species, ratio))

col_headers = ["Species", "Avg petal length", "Max petal width ", "Width/length ratio"]
print(f"{col_headers[0]:<15} {col_headers[1]:<20} {col_headers[2]:<20} {col_headers[3]:<20}")
print(f"{list_of_species[0].title():<15} {avg_measurements_per_species[0][1]:^20} {max_petal_width_per_species[0][1]:^20} {petal_length_to_width_ratio_per_species[0][1]:^20}")
print(f"{list_of_species[1].title():<15} {avg_measurements_per_species[1][1]:^20} {max_petal_width_per_species[1][1]:^20} {petal_length_to_width_ratio_per_species[1][1]:^20}")
print(f"{list_of_species[2].title():<15} {avg_measurements_per_species[2][1]:^20} {max_petal_width_per_species[2][1]:^20} {petal_length_to_width_ratio_per_species[2][1]:^20}")
```
