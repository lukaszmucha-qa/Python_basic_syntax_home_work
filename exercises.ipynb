```python
# Zadanie 1 - Kalkulator BMI
waga = float(input("Podaj swoją wagę w kilogramach: "))
wzrost = float(input("Podaj swój wzrost w metrach: "))
bmi = waga / (wzrost ** 2)
print(f"BMI: {bmi:.2f}")
if bmi < 18.5:
    print("Niedowaga")
elif 18.5 <= bmi < 24.9:
    print("Waga prawidłowa")
elif 25 <= bmi < 29.9:
    print("Nadwaga")
else:
    print("Otyłość")


```
