```python
# Zadanie 1 - Kalkulator BMI
weigh = float(input("Podaj swoją wagę w kilogramach: "))
height = float(input("Podaj swój wzrost w metrach: "))
bmi = weigh / (height ** 2)
print(f"BMI: {bmi:.2f}")
if bmi < 18.5:
    print("Niedowaga")
elif 18.5 <= bmi < 24.9:
    print("Waga prawidłowa")
elif 25 <= bmi < 29.9:
    print("Nadwaga")
else:
    print("Otyłość")


```

```python
# Zadanie 2 - Konfersja temperatur
temp_c = float(input("Podaj temperaturę w stopniach Celsjusza: "))
temp_f = (temp_c * 9/5) + 32
print(f"Temperatura w stopniach Fahrenheita: {temp_f:.2f}")
temp_k = temp_c + 273.15
print(f"Temperatura w stopniach Kelvina: {temp_k:.2f}")
print(f"{temp_c:.2f}°C = {temp_f:.2f}°F = {temp_k:.2f}K")
```

```python
# Zadanie 3 - Podział zbioru dany
number_of_samples = 1000
train_ratio = 0.7
validation_ratio = 0.15
test_ratio = 0.15
train_size = int(number_of_samples * train_ratio)
validation_size = int(number_of_samples * validation_ratio)
test_size = int(number_of_samples * test_ratio)
print(f"Rozmiar zbioru treningowego: {train_size}")
print(f"Rozmiar zbioru walidacyjnego: {validation_size}")
print(f"Rozmiar zbioru testowego: {test_size}")
print(f"Suma rozmiarów = 1000 is: " + str(train_size + validation_size + test_size == number_of_samples))
```
